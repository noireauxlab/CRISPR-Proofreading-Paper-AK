```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b = 1
a = [10, 1000, 100000, 10000000]
c = 0
shift1 = 9
shift2 = 16
Ew1 = 7
Ew2 = 7
Ebp = -1.3
end = 24
Estep = -5
ps = []
for j in range(len(a)):
    probs = []
    for i in range(end):
        if i in range(0, shift1):
            Et = i*Ebp
        elif i in range(shift1, shift2):
            Et = Ew1 + i*Ebp
        else:
            Et = Ew1 + Ew2 + i*Ebp
        p = b/(1+a[j]*np.exp(Et)) + c
        probs.append(p)
    ps.append(probs)
```

```python
ax = plt.axes()
plt.rcParams['figure.figsize'] = [20, 13]
for j in range(len(ps)):
    plt.plot(range(end), ps[j], label = a[j], lw = 5)
plt.ylabel('Binding Probability', fontsize = 50)
plt.xlabel('Guide RNA length (bp)', fontsize = 50)
plt.axvspan(8.5, 15.5, facecolor='silver', alpha=0.5)
ax.set_title('dCas9 CRISPRa ' + r'$\alpha$' + ' dependence', fontsize = 50)
plt.axvspan(15.5, 26, facecolor='grey', alpha=0.5)
plt.xlim(-0.2, 21)
plt.xticks(np.arange(0, 21, 1),  fontsize = 40)
plt.yticks(fontsize = 50)
sns.set_style("whitegrid")
```
